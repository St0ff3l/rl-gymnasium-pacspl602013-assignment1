## 1. Environment Selection

For this assignment, I selected the **CarRacing-v3** environment from the Box2D category in the Gymnasium library.

* **Environment Description:** A top-down racing environment where the agent learns to drive from pixel inputs (96x96 RGB images). The track is randomly generated every episode.
* **Reason for Selection:** CarRacing is a more advanced benchmark that requires the reinforcement learning agent to simultaneously process visual inputs (via Convolutional Neural Networks) and output complex control decisions (steering, gas, braking).

## 2. Training the Agent

**Personal Note:** In this stage, I utilized the `PPO` (Proximal Policy Optimization) algorithm from the `stable-baselines3` library. PPO is known for being a stable and highly efficient policy gradient method. To allow the agent to learn basic driving behaviors from the pixel inputs, the model was set to train for 100,000 timesteps. As shown in the training logs below, we can observe the model's learning progress.

```python
import gymnasium as gym
from stable_baselines3 import PPO

# 1. Initialize the environment
# Setting continuous=False simplifies the action space to 5 discrete actions, 
# which helps the model converge faster during early training.
env = gym.make("CarRacing-v3", continuous=False)

# 2. Instantiate the PPO model using a Convolutional Neural Network (CNN) policy
# Since the observation space consists of pixel arrays (images), a CnnPolicy is required.
model = PPO("CnnPolicy", env, verbose=1)

# 3. Start training
print("Training started...")

# For the purpose of validating the pipeline for the assignment, 
# 20,000 timesteps is sufficient to generate the necessary training logs.
model.learn(total_timesteps=100000) 

print("Training completed!")

# 4. Save the trained model
model.save("ppo_carracing_model")
print("Model saved successfully!")
env.close()
```

## 3. Testing and Rendering

**Personal Note:** Following the training phase, I instantiated a new test environment with `render_mode="human"` to visually evaluate the agent's performance. By having the trained convolutional neural network predict actions based on the pixel observations, we can intuitively observe the car's driving behavior and its ability to navigate the randomly generated track.

```python
import gymnasium as gym
from stable_baselines3 import PPO

model = PPO.load("ppo_carracing_model")

# 4. Test the model and render the animation
test_env = gym.make("CarRacing-v3", continuous=False, render_mode="human")
obs, info = test_env.reset()

for i in range(1000):
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = test_env.step(action)
    
    if terminated or truncated:
        obs, info = test_env.reset()

test_env.close()
```
