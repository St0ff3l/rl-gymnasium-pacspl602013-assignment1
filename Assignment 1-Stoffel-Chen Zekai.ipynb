## 1. Environment Selection

For this assignment, I selected the **Pendulum-v1** environment from the Gymnasium library.

* **Environment Description:** The system consists of a pendulum attached at one end to a fixed point. The starting state is a random angle in [-pi, pi] and a random angular velocity in [-1, 1]. The goal is to apply torque to swing it into an upright position. The episode truncates at 200 time steps.

* **Reason for Selection:** Pendulum-v1 is a classic problem in control theory. It has a continuous action space (torque) and a continuous observation space (x-y coordinates and angular velocity). Its reward function is defined as $r = -(\theta^2 + 0.1 \cdot \theta_{dt}^2 + 0.001 \cdot \text{torque}^2)$. This setup makes it a great choice for testing the SAC algorithm.

```python
import gymnasium as gym
from stable_baselines3 import SAC

# Initialize the Pendulum environment
env = gym.make("Pendulum-v1")

# Instantiate the SAC model
model = SAC("MlpPolicy", env, verbose=1)
```

## 2. Training the Agent

I used the `SAC` (Soft Actor-Critic) algorithm from the `stable_baselines3` library. SAC is an off-policy maximum entropy deep reinforcement learning algorithm. It is highly efficient for environments with continuous action spaces. The agent was trained for 20,000 timesteps, which is sufficient to learn the pendulum swing-up task.

```python
# Start training
print("Training started...")

# Train the agent for 20,000 timesteps
model.learn(total_timesteps=20000)

print("Training completed!")

# Save the trained model
model.save("sac_pendulum_model")
print("Model saved successfully!")

env.close()
```

## 3. Testing and Rendering

I test the agent using render_mode="human". The trained SAC model outputs the torque. This allows me to visually check if the pendulum swings up and keeps its balance.

```python
import gymnasium as gym
from stable_baselines3 import SAC

# Load the trained model and render the animation
model = SAC.load("sac_pendulum_model")
test_env = gym.make("Pendulum-v1", render_mode="human")
obs, info = test_env.reset()

for i in range(5000):
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = test_env.step(action)
    
    if terminated or truncated:
        obs, info = test_env.reset()

test_env.close()
```
